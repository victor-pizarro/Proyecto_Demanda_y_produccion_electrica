```python
train_01 = df_demanda.loc[df_demanda_01.index < '2021-01-01 00:00:00']
test_01 = df_demanda.loc[df_demanda_01.index >= '2021-01-01 00:00:00']

fig, ax = plt.subplots(figsize=(15,5))
train_01.plot(ax=ax, label='Training Set')
test_01.plot(ax=ax, label='Test Set')
ax.axvline('2021-01-01 00:00:00', color='black', ls='--')
plt.legend(['Training Set', 'Test Set'])
plt.ylabel('Demanda (MWh)')
plt.grid(axis='x')
plt.savefig("images/Demanda_electrica_train_test_split.jpg",bbox_inches='tight',dpi=300)
plt.show()
```

```python
datos_serie_temporal(train_01);
datos_serie_temporal(test_01);
festivos_y_fin_de_semana(train_01);
festivos_y_fin_de_semana(test_01);

predictoras=['Año', 'Mes', 'Dia_año', 'Dia_semana', 'Hora', 'Cuartos', 'Semana_año', 'fest_nacional', 'fin_de_semana']
target=['Demanda']

X_train_01 = train_01[predictoras]
y_train_01 = train_01[target]
X_test_01 = test_01[predictoras]
y_test_01 = test_01[target]

print(f'''X_Train {X_train_01.shape}
X_test {X_test_01.shape}
y_train {y_train_01.shape}
y_test {y_test_01.shape}''')
```

`FEATURE SCALER`

```python
minmax = MinMaxScaler()
X_train_escalado_01 = minmax.fit_transform(X_train_01)
X_test_escalado_01 = minmax.transform(X_test_01)
```

**`Hyperparameters XGBRegressor`**

- `max_depth`: Máxima profundidad del arbol. Se usa para controlar el overfitting. A mayor profundidad, mayor overfitting.


- `n_estimators`: Número de gradient boosted trees, equivalente al numero de boosting rounds.


- `min_child_weight`: Peso mínimo necesario para crear un nuevo nodo en el árbol. Un tamaño más pequeño permite que el algoritmo cree elementos secundarios que correspondan a menos muestras lo que permite arboles más complejos. A menos min_child_weight, mayor overfitting.


- `eta` or `learning_rate`: parámetro que controla la tasa de aprendizaje. Corresponde a la reducción de los pesos asociados a las caracteristicas despues de cada ronda, define la cantidad de 'corrección' que se hace en cada paso. Un valor bajo hace que nuestro modelo sea más robusto al overfitting, cuanto menor sea nuestra `learning rate` mejor.

`GRID SEARCH CROSS VALIDATION`

```python
gs_cv = GridSearchCV(
    estimator = XGBRegressor(),
    param_grid = {
        'n_estimators':[1000],
        'max_depth':[2,3], #5,6
        'min_child_weight':[50,60], # 5,7,9
        'learning_rate':[0.01,0.05]}, #0.05, 0.1
    cv = 6,
    return_train_score = True,
    scoring = 'neg_mean_squared_error')

grid_result = gs_cv.fit(X_train_escalado_01, y_train_01)
grid_result
```

```python
grid_result.best_estimator_
grid_result.best_params_
```

`XGBRegressor model`

```python
xgbr = XGBRegressor(n_estimators=1000,
                    learning_rate=0.01, 
                    max_depth=2,
                    min_child_weight=7)
xgbr
```

```python
xgbr.fit(X_train_escalado_01, y_train_01,
       eval_set=[(X_train_escalado_01, y_train_01), (X_test_escalado_01, y_test_01)],
       verbose=10)
```

```python
prediction_train_01 = xgbr.predict(X_train_escalado_01)
prediction_test_01 = xgbr.predict(X_test_escalado_01)
```

```python
df_demanda_02 = df_demanda.copy()
df_demanda_02
```

```python
train_02 = df_demanda.loc[df_demanda_02.index < '2021-01-01 00:00:00']
test_02 = df_demanda.loc[df_demanda_02.index >= '2021-01-01 00:00:00']

fig, ax = plt.subplots(figsize=(15,5))
train_02.plot(ax=ax, label='Training Set')
test_02.plot(ax=ax, label='Test Set')
ax.axvline('2021-01-01 00:00:00', color='black', ls='--')
plt.legend(['Training Set', 'Test Set'])
plt.grid(axis='x')
plt.show()
```

```python
datos_serie_temporal(train_02);
datos_serie_temporal(test_02);
festivos_y_fin_de_semana(train_02);
festivos_y_fin_de_semana(test_02);
```

```python
predictoras=['Año', 'Mes', 'Dia_año', 'Dia_semana', 'Hora', 'Cuartos', 'Semana_año', 'fest_nacional', 'fin_de_semana']
target=['Demanda']

X_train_02 = train_02[predictoras]
y_train_02 = train_02[target]
X_test_02 = test_02[predictoras]
y_test_02 = test_02[target]

print(f'''X_Train {X_train_02.shape}
X_test {X_test_02.shape}
y_train {y_train_02.shape}
y_test {y_test_02.shape}''')
```

```python
minmax = MinMaxScaler()
X_train_escalado_02 = minmax.fit_transform(X_train_02)
X_test_escalado_02 = minmax.transform(X_test_02)
```

`Hyperparameters RandomForestRegressor`:

- `n_estimators`: Número de arboles (trees) en el bosque (forest)


- `max_features`: Número máximo de features a considerar para realizar el split del nodo.


- `max_depth`: Número máximo de splis (profundidad máxima del árbol)

- `min_samples_split`: Número mínimo de datos en un nodo antes de hacer el split.

- `min_samples_leaf`: Número mínimo de datos permitidos en una hoja.

- `bootstrap`: metodo para muestrear puntos de datos (con o sin reemplazo)


```python
gs_cv = GridSearchCV(
    estimator = RandomForestRegressor(),
    param_grid = {
        'n_estimators':[100],
        'max_depth':[6,7,10],
        'max_features':[5,7],
        'min_samples_split':[3,5]},
    cv = 5,
    return_train_score = True,
    scoring = 'neg_mean_squared_error')

grid_result = gs_cv.fit(X_train_escalado_02, y_train_02)
grid_result
```

```python
grid_result.best_estimator_
grid_result.best_params_
```

```python
rfr = RandomForestRegressor(n_estimators=100,
                            max_depth=9,
                            min_samples_split=5,
                            max_features=5)

rfr.fit(X_train_escalado_02, y_train_02)
```

```python
prediction_train_02 = rfr.predict(X_train_escalado_02)
prediction_test_02 = rfr.predict(X_test_escalado_02)
```

```python
df_metrics_rfr = metricas(y_train_02['Demanda'], y_test_02['Demanda'], prediction_train_02, prediction_test_02, 'RandomForestRegressor')
df_metrics_rfr
```

```python
feature_importances= pd.DataFrame(data=rfr.feature_importances_,
                                  index=X_train_02.columns,
                                  columns=['feat_importances'])

feature_importances.sort_values('feat_importances').plot(kind='barh',figsize=(10,5))
plt.legend(['importance'],bbox_to_anchor=(1.0,0.15))
plt.title('Random Forest Regressor')
plt.grid(axis='x')
plt.xlabel('Feature importance')
plt.ylabel('Feature')
plt.show()
```
