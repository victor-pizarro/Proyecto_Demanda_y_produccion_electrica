```python
# TRAIN TEST SPLIT
train_0X = df_demanda_0X.loc[df_demanda_0X.index < '2020-01-01 00:00:00']
test_0X = df_demanda_0X.loc[df_demanda_0X.index >= '2021-01-01 00:00:00']
```

```python
fig, ax = plt.subplots(figsize=(25,5))
train_0X.plot(ax=ax, label='Training Set', )
test_0X.plot(ax=ax, label='Test Set')
ax.axvline('2021-01-01 00:00:00', color='black', ls='--')
ax.axvline('2020-01-01 00:00:00', color='black', ls='--')
plt.legend(['Training Set', 'Test Set'])
plt.ylabel('Demanda (MWh)')
plt.grid(axis='x')
plt.savefig("images/Train_test_split_no_covid.jpg",bbox_inches='tight',dpi=300)
plt.show()
```

```python
# FEATURE ENGINEERING
datos_serie_temporal(train_0X);
datos_serie_temporal(test_0X);
festivos_y_fin_de_semana(train_0X);
festivos_y_fin_de_semana(test_0X);
```

```python
predictoras=['Año', 'Mes', 'Dia_año', 'Dia_semana', 'Hora', 'Cuartos', 'Semana_año', 'fest_nacional', 'fin_de_semana']
target=['Demanda']

X_train_0X = train_0X[predictoras]
y_train_0X = train_0X[target]
X_test_0X = test_0X[predictoras]
y_test_0X = test_0X[target]

print(f'''X_Train {X_train_0X.shape}
X_test {X_test_0X.shape}
y_train {y_train_0X.shape}
y_test {y_test_0X.shape}''')
```

```python
# FEATURE SCALING
minmax = MinMaxScaler()
X_train_escalado_0X = minmax.fit_transform(X_train_0X)
X_test_escalado_0X = minmax.transform(X_test_0X)
```

```python
# GRID SEARCH CROSS VALIDATION
gs_cv = GridSearchCV(
    estimator = XGBRegressor(),
    param_grid = {
        'n_estimators':[500,1000],
        'max_depth':[3,5,7],
        'min_child_weight':[50,60],
        'learning_rate':[0.01,0.05]},
    cv = 5,
    return_train_score = True,
    scoring = 'neg_mean_squared_error')

grid_result = gs_cv.fit(X_train_escalado_0X, y_train_0X)
grid_result
```

```python
# GRID SEARCH RESULTS
grid_result.best_estimator_
grid_result.best_params_
```

```python
# MODELO
xgbr = XGBRegressor(n_estimators=500,
                    learning_rate=0.01, 
                    max_depth=7,
                    min_child_weight=50)
xgbr
```

```python
# FIT
xgbr.fit(X_train_escalado_0X, y_train_0X,
       eval_set=[(X_train_escalado_0X, y_train_0X), (X_test_escalado_0X, y_test_0X)],
       verbose=10)
```

```python
# PREDICCION
prediction_train_0X = xgbr.predict(X_train_escalado_0X)
prediction_test_0X = xgbr.predict(X_test_escalado_0X)
```

`FEATURE SCALER`

```python
minmax = MinMaxScaler()
X_train_escalado_01 = minmax.fit_transform(X_train_01)
X_test_escalado_01 = minmax.transform(X_test_01)
```

**`Hyperparameters XGBRegressor`**

- `max_depth`: Máxima profundidad del arbol. Se usa para controlar el overfitting. A mayor profundidad, mayor overfitting.


- `n_estimators`: Número de gradient boosted trees, equivalente al numero de boosting rounds.


- `min_child_weight`: Peso mínimo necesario para crear un nuevo nodo en el árbol. Un tamaño más pequeño permite que el algoritmo cree elementos secundarios que correspondan a menos muestras lo que permite arboles más complejos. A menos min_child_weight, mayor overfitting.


- `eta` or `learning_rate`: parámetro que controla la tasa de aprendizaje. Corresponde a la reducción de los pesos asociados a las caracteristicas despues de cada ronda, define la cantidad de 'corrección' que se hace en cada paso. Un valor bajo hace que nuestro modelo sea más robusto al overfitting, cuanto menor sea nuestra `learning rate` mejor.

`GRID SEARCH CROSS VALIDATION`

```python
gs_cv = GridSearchCV(
    estimator = XGBRegressor(),
    param_grid = {
        'n_estimators':[1000],
        'max_depth':[2,3], #5,6
        'min_child_weight':[50,60], # 5,7,9
        'learning_rate':[0.01,0.05]}, #0.05, 0.1
    cv = 6,
    return_train_score = True,
    scoring = 'neg_mean_squared_error')

grid_result = gs_cv.fit(X_train_escalado_01, y_train_01)
grid_result
```
